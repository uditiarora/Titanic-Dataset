```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
```

```python
data = pd.read_csv("train.csv")
data.drop("Embarked",axis=1,inplace=True)
data.Age.fillna(data.Age.mean(),inplace=True)
temp = np.zeros(len(data))
for i in range(len(data)):
    if(data["Age"][i]<=16):
        temp[i]=3
    else:
        if(data["Sex"][i]=="male"):
            temp[i]=1
        else:
            temp[i]=2
data["female_male_child"] = temp
data.drop("Sex",inplace=True,axis=1)
data.drop("Cabin",inplace=True,axis=1)
data.drop("Ticket",inplace=True,axis=1)
data.drop("Name",inplace=True,axis=1)
y_train = np.array(data["Survived"])
data.drop("Survived",inplace=True,axis=1)
x_train = np.array(data)
data.describe()
```

```python
data = pd.read_csv("test.csv")
data.drop("Embarked",axis=1,inplace=True)
data.Age.fillna(data.Age.mean(),inplace=True)
temp = np.zeros(len(data))
for i in range(len(data)):
    if(data["Age"][i]<=16):
        temp[i]=3
    else:
        if(data["Sex"][i]=="male"):
            temp[i]=1
        else:
            temp[i]=2
data["female_male_child"] = temp
data.drop("Sex",inplace=True,axis=1)
data.drop("Cabin",inplace=True,axis=1)
data.drop("Ticket",inplace=True,axis=1)
data.drop("Name",inplace=True,axis=1)
data.Fare.fillna(data.Fare.mean(),inplace=True)
x_test = np.array(data)
data.describe()
```

```python
alg1 = RandomForestClassifier()
alg1.fit(x_train,y_train)
y_pred = alg1.predict(x_test).reshape(-1,1)
temp = np.array(data["PassengerId"]).reshape(-1,1)
arr = np.concatenate((temp,y_pred),axis=1)
np.savetxt("pred.csv",arr,fmt="%d",delimiter=",")
```
